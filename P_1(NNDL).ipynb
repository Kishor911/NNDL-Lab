```python
import numpy as np
m1=np.array([[1,2,3], [4,5,6]])
m2=np.array([[7,8], [10,11], [12,13]])

print("MatrixA", m1)
print()
print("MatrixB", m2)
print()

res=np.matmul(m1,m2)
print("Matrix Multiplication", res)
```
